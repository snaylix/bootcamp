Import packages

```python
import requests
import re
```

Make a URL request

```python
url = "https://www.lyrics.com/artist/Helge-Schneider/748954"
param_dict = {'db':'pubmed', 'term':'escherichia', 'rettype':'uilist'}

r_helge = requests.get(url, params=param_dict)
```

Check the status code in order to assess whether our request was successful

```python
r_helge.status_code
```

It is a requests type

```python
type(r_helge)
```

Return HTML file of the website as a string

```python
print(r_helge.text)
```

```python
type(r_helge.text)
```

Save the html in a txt file and search in an editor

```python
with open('_RES/Helge.txt', 'w') as file:
    file.write(r_helge.text)
```

```python
# <a href="/lyric/10047067/Helge+Schneider/Texas">Texas</a>
```

```python
pattern = "lyric/\d+/Helge\+Schneider/[^\"]+"
```

```python
songs_helge = re.findall(pattern, r_helge.text, re.IGNORECASE)
```

```python
songs_helge
```

Loop that goes through all song URLs that you collected previously and construct a complete URL

```python
for song in range(len(songs_helge)):
    songs_helge[song] = 'https://www.lyrics.com/' + songs_helge[song]
```

```python
songs_helge
```
